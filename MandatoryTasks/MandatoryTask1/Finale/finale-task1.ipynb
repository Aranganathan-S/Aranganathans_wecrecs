```python
from transformers import pipeline
```

I used to pretrain models to handle the question answering and at the same time to translate english texts to french

```python
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
translation_pipeline = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
```

```python
context = """
Harry Potter is a series of seven fantasy novels written by British author J.K. Rowling.
The novels chronicle the lives of a young wizard, Harry Potter, and his friends Hermione Granger and Ron Weasley,
all of whom are students at Hogwarts School of Witchcraft and Wizardry.
"""

questions = [
    "Who wrote the novel Harry Potter?",
    "Where did Harry Potter study?",
    "Name one of Harry Potter's friends."
]
```

```python
def english_to_french_qa(question, context):
    qa_result = qa_pipeline(question=question, context=context)
    english_answer = qa_result["answer"]
    french_translation = translation_pipeline(english_answer)[0]["translation_text"]
    return english_answer, french_translation
```

```python
for i, q in enumerate(questions, 1):
    eng_ans, fr_ans = english_to_french_qa(q, context)
    print(f"Question {i}: {q}")
    print(f"English Answer : {eng_ans}")
    print(f"French Answer  : {fr_ans}")
```
